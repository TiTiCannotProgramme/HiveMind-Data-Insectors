```python
import torch, torchvision
import sys # Python system library needed to load custom functions
import math # module with access to mathematical functions
import os # for changing the directory

import numpy as np  # for performing calculations on numerical arrays
import pandas as pd  # home of the DataFrame construct, _the_ most important object for Data Science
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt  # allows creation of insightful plots

sys.path.append('../audio_preprocessing')
sys.path.append('../src')
sys.path.append('../model_training_utils')


from gdsc_utils import PROJECT_DIR
import model_eval

os.chdir(PROJECT_DIR) # changing our directory to root
```

```python
import json

with open('audio_preprocessing/saved_data/upsampled_data_size_128_512_train_and_val.json') as f:
    my_info = json.load(f)

mean, std, class_weights = my_info["mean"], my_info["std"], my_info["weights"]
```

```python
resnet_model = torch.load('models/resnet34_train_and_val/resnet_model_epoch_13.pth')
regnet_model = torch.load('models/regnet_y_3_2gf_train_and_val/regnet_model_epoch_12.pth')
dense_net_169_model = torch.load('models/DenseNet169_train_and_val/densenet_169_model_epoch_11.pth')
shuffle_net_model = torch.load('models/ShuffleNet_train_and_val/shuffle_net_model_epoch_10.pth')
resnext_50_model = torch.load('models/resnext50_train_and_val/resnext_model_epoch_12.pth')
efficient_net_v1_model = torch.load('models/EfficientNet_train_and_val/efficient_model_epoch_11.pth')
```

```python
# generate file paths in the test data folder

test_paths = [] 
test_labels = [] 
test_file_names = []
stem = "data/test"
for i in range(556):
    test_paths.append(f"{stem}/{i}.wav")
    test_file_names.append(f"{i}.wav")
    test_labels.append(0)
```

```python
test_result = model_eval.evaluate_audio_classes(
    [regnet_model, dense_net_169_model, 
     shuffle_net_model, resnext_50_model, efficient_net_v1_model], 
    test_paths, test_labels, mean, std
)
```

```python
submit_format_test_result_dict = {
    "file_name": test_file_names,
    "predicted_class_id": test_result["predicted_class_id"]
}

df_test_result = pd.DataFrame.from_dict(submit_format_test_result_dict)
df_test_result.head()
```

```python
save_name = "reg_efficient_v1_shuffle_dense_resnext_combined"

df_test_result.to_csv(f"test_results/{save_name}.csv", index=False)
```
